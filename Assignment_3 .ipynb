```python
df_table=df.iloc[:,1:6]
df_table
```

```python
val=stats.chi2_contingency(df_table)
val
```

```python
type(val)
```

```python
no_of_rows=len(df_table.iloc[0:2,0])
no_of_columns=len(df_table.iloc[0,0:4])
df=(no_of_rows-1)*(no_of_columns-1)
print("Degree of freedom=",df)
```

```python
Expected_value=val[3]
Expected_value
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2/e for o,e in zip(df_table.values,Expected_value)])
chi_square_statistic=chi_square[0]+chi_square[1]
chi_square_statistic
```

```python
critical_value=chi2.ppf(0.95,3)
critical_value
```

```python
if chi_square_statistic >= critical_value:
  print('Dependent (reject the null hypothesis H0)')
else:
    print('Independent (fail to reject HO)')
```

```python
p_value=1-chi2.cdf(chi_square_statistic,3)
p_value
```

```python
if p_value <=0.05:
  print('Dependent (reject H0)')
else:
  print('Independent (fail to reject H0)')
```
